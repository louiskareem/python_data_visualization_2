```python
from matplotlib import pyplot as plt
```

```python
import pandas as pd
```

```python
data = pd.read_csv('cgpa.csv')
data = data.head(30)

# plt.plot(data['rollno'], data['cgpa'], color='purple', label='Line Graph') 
# plt.scatter(data['rollno'], data['cgpa'], color='red', label='Scatter')
plt.bar(data['rollno'], data['cgpa'], color=['green', 'blue'], alpha=.7, label = 'Bar Graph') # barh displays horizontal bar graph
plt.legend()
plt.xlabel('Roll No', color='green')
plt.ylabel('CGPA', color='blue')
plt.title('CGPA vs Roll No', color='brown')
plt.show()
```

```python
data1 = pd.read_csv('cgpa.csv')
data = data1.head(30)

fig = plt.figure()
fig.patch.set_facecolor('grey')

graph1 = fig.add_subplot(2,2,1, facecolor='red')
graph1.plot(data['rollno'], data['cgpa'])
graph1.set_title('CGPA vs Roll No')

graph2 = fig.add_subplot(2,2,2, facecolor='yellow')
graph2.plot(data['cgpa'], data['rank'])
graph2.set_title('Rank vs CGPA')

graph3 = fig.add_subplot(2,2,3, facecolor='lightgrey')
x = len(data1[data1.cgpa>=9])
x1 = len(data1[(data1.cgpa>= 8) & (data1.cgpa< 9)])
x2 = len(data1[data1.cgpa< 8])
graph3.axis('equal')
graph3.pie([x,x1,x2], colors=['Yellow', 'Red', 'Orange'], labels = ['9 Pointer', '8 Pointer', 'Others'])
graph3.legend(title = 'Description')

plt.show()
```
